```python
# Conectamos con la base de datos chinook.db
connection = sqlite3.connect("./chinook.db")

# Obtenemos un cursor que utilizaremos para hacer las queries
crsr = connection.cursor()
```

```python
# Con esta función leemos los datos y lo pasamos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)
```

```python
# 1. Facturas de Clientes de Brasil, Nombre del cliente, Id de factura, fecha de la factura y el pais de la factura* 

query = """ 

SELECT b.FirstName || ' ' || b.LastName AS "Contacto", b.Company, a.InvoiceId As "ID Facturas",a.InvoiceDate, b.Country
FROM invoices AS a
INNER JOIN customers AS b
ON a.CustomerId = b.CustomerId

WHERE b.Country = "Brazil"
LIMIT 5

"""

sql_query(query)
```

```python
# 2. Facturas de Clientes de Brasil*

query = """ 

SELECT b.Company, a.InvoiceId As "ID Facturas",a.InvoiceDate, b.Country

FROM invoices AS a
INNER JOIN customers AS b
ON a.CustomerId = b.CustomerId 

LEFT JOIN invoice_items as c
ON a.InvoiceId = c.InvoiceId

WHERE b.Country = "Brazil"

LIMIT 5
"""

sql_query(query)
```

```python
# 3. Muestra cada factura asociada a cada agente de ventas con su nombre completo*

query = """ 

SELECT a.InvoiceId,d.FirstName,d.LastName

FROM invoices AS a
INNER JOIN customers AS b
ON a.CustomerId = b.CustomerId 

INNER JOIN invoice_items as c
ON a.InvoiceId = c.InvoiceId

INNER JOIN employees as d
ON b.CustomerId = d.EmployeeId

WHERE d.Title LIKE "%Sales%"
GROUP BY a.InvoiceId

LIMIT 10
"""
sql_query(query)
```

```python
# 4. Muestra el nombre del cliente, el pais, el nombre del agente y el total 

query = """ 

SELECT c.FirstName || ' ' || c.LastName AS CustomerName, 
       c.Country,
       e.FirstName || ' ' || e.LastName AS EmployeeName, 
       SUM(ii.Quantity) AS TotalItemsSold
FROM invoices i
JOIN customers c ON i.CustomerId = c.CustomerId
JOIN invoice_items ii ON i.InvoiceId = ii.InvoiceId
JOIN employees e ON c.SupportRepId = e.EmployeeId
GROUP BY e.EmployeeId, c.CustomerId, c.FirstName, c.LastName, c.Country, e.FirstName, e.LastName

LIMIT 10

"""
sql_query(query)
```

```python
# 5. Muestra cada articulo de la factura con el nombre de la cancion


query = """ 

SELECT e.Name, c.InvoiceId

FROM invoices AS a
JOIN customers AS b
ON a.CustomerId = b.CustomerId 

INNER JOIN invoice_items as c
ON a.InvoiceId = c.InvoiceId

INNER JOIN employees as d
ON b.SupportRepid = d.EmployeeId

INNER JOIN tracks as e
ON c.TrackiD = e.Trackid


LIMIT 5
"""
sql_query(query)
```

```python
# 6. Muestra todas las conciones con su nombre, formato, album y genero 

query = """ 

SELECT t.Name AS Cancion, mt.Name as Formato, a.Title as Album, g.Name as Género
FROM invoice_items as ii

JOIN tracks as t
ON ii.TrackId = t.TrackId

JOIN albums as a
ON t.AlbumId = a.AlbumId

JOIN media_types as mt
ON t.MediaTypeId = mt.MediaTypeId

JOIN genres as g
ON t.GenreId = g.GenreId 

LIMIT 10

"""
sql_query(query)
```

```python
# 7. Muestra cuantas canciones hay en cada playlist y el nombre de cada playlist


query = """ 

SELECT COUNT(t.Name) as "N° Canciones", pl.Name as "Lista de Reproducción"
FROM playlist_track as pt
JOIN tracks as t
ON t.TrackId = pt.TrackId
JOIN playlists as pl
ON pt.PlaylistId = pl.PlaylistId

GROUP BY pl.Name
"""
sql_query(query)
```

```python
# 8. Muestra cuánto ha vendido cada empleado

query = """ 

SELECT e.FirstName || " " || e.LastName as Vendedor, SUM(ii.UnitPrice * ii.Quantity) AS "Ventas X  Vendedor"
FROM invoices i
JOIN customers c ON i.CustomerId = c.CustomerId
JOIN invoice_items ii ON i.InvoiceId = ii.InvoiceId
JOIN employees e ON c.SupportRepId = e.EmployeeId
WHERE e.Title LIKE "%Sales%"
GROUP BY e.EmployeeId, e.FirstName, e.LastName;


"""
sql_query(query)
```

```python
# 9. Quien ha sido el agente de ventas que más ha vendido en 2009? 

query = """ 

SELECT e.FirstName || ' ' || e.LastName AS "Nombre Vendedor", SUM(ii.UnitPrice * ii.Quantity) AS "Venta por Vendedor"
FROM invoices i
JOIN customers c ON i.CustomerId = c.CustomerId
JOIN invoice_items ii ON i.InvoiceId = ii.InvoiceId
JOIN employees e ON c.SupportRepId = e.EmployeeId
WHERE e.Title LIKE '%Sales%' AND strftime('%Y', i.InvoiceDate) = '2009'
GROUP BY e.EmployeeId



"""
sql_query(query)
```

```python
# 10. Cuáles son las 3 bandas que más han vendido? 

query = """ 

SELECT a.Name AS "Banda", SUM(ii.UnitPrice * ii.Quantity) AS "Ventas Totales Banda"
FROM artists a
JOIN albums al ON a.ArtistId = al.ArtistId
JOIN tracks t ON al.AlbumId = t.AlbumId
JOIN invoice_items ii ON t.TrackId = ii.TrackId
GROUP BY a.ArtistId
ORDER BY "Ventas Totales Banda" DESC
LIMIT 3;


"""
sql_query(query)
```
