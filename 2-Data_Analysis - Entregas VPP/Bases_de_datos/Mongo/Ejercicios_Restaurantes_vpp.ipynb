```python
# 6. Encontrar los restaurantes que no preparan ninguna cocina de 'estadounidense' y lograron una puntuacin superior a 70 y se ubicaron en una longitud inferior a -65.754168. Nota: Realice esta consulta sin usar el operador $and

projection = {"_d":0}
filter = {"cuisine": {"$ne": "American "},
          "grades.score": {"$gt":80},
          "address.coord.0":{"$lt": -65.754168 }
          
          }

list(restaurantes.find(projection = projection, filter=filter).limit(5))
```

```python
# 7. Devolver los restaurantes que no preparan cocina americana y lograron un punto de calificacin 'A' que no pertenece al distrito de Brooklyn. El documento debe mostrarse segn la cocina en orden descendente.

projection = {"_id":0}
filter = { "cousine": {"$ne": "American " }, "grades.grade": "A", "borough": {"$ne":"Brooklyn"}
}

list(restaurantes.find(filter=filter, projection = projection).sort({"cousine":-1}))
```

```python
# 8. Devolver los restaurantes que pertenecen al distrito Bronx y preparan platos americanos o chinos

projection = {"_id":0}
filter = {"borough": "Bronx", 
          "cuisine":{"$in": ["American ", "Chinese"]}
          }

list(restaurantes.find(filter=filter, projection = projection))
```

```python
# 9. Devolver ID del restaurante, nombre, distrito y la cocina para aquellos restaurantes que pertenecen al distrito de Staten Island o Queens o Bronx o Brooklyn

projection = {"_id":0,"restaurant_id":1,"name":1, "borough":1, "cuisine":1}
filter = {"borough":{"$in":["State island", "Queens", "Brooklyn"] }}

list(restaurantes.find(filter=filter, projection = projection))
```

```python
# 10. Devolver ID del restaurante, nombre, distrito y la cocina de aquellos restaurantes que lograron una puntuacin que no supere los 10

projection = {"_id":0,"restaurant_id":1,"name":1, "borough":1, "cuisine":1}
filter = {"grades.score": {"$gt":10}}

list(restaurantes.find(filter=filter, projection = projection).limit(5))
```

```python
# 11. Devolver ID del restaurante, el nombre y las calificaciones del restaurante para aquellos restaurantes que obtuvieron una calificacin de "A" 
# y obtuvieron un puntaje de 11 en una fecha ISO "2014-08-11T00: 00: 00Z" entre muchas fechas de encuesta

from datetime import datetime

projection = {"_id":0,"restaurant_id":1,"name":1, "grades":1}
filter = { "grades": {
         "$elemMatch":{
              "grade":"A", 
              "score":11,
              "date": datetime(2014,8,11)
              
          }}}

list(restaurantes.find(filter=filter, projection = projection))
```

```python
# 12. Devolver ID del restaurante, nombre, direccin y ubicacin geogrfica del restaurante de aquellos 
# donde el segundo elemento de la matriz coord contiene un valor que es ms de 42 y hasta 52

projection = {"_id":0,"restaurant_id":1,"name":1, "address":1 }
filter = { "address.coord.1": {"$gt":42,
                              "$lte": 52}}

list(restaurantes.find(filter=filter, projection = projection).limit(5))
```
