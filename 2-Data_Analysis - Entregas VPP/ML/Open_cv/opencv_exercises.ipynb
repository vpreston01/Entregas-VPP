```python
blank = np.zeros(img1.shape, dtype='uint8')
cv.imshow('Blank', blank)

# Draw a white rectangle on the blank image
rectangle = cv.rectangle(blank.copy(), (200, 100), (500, 300), 255, -1)
cv.imshow("Rectangle", rectangle)

# Invert the colors of the rectangle
mask = cv.bitwise_not(rectangle)
cv.imshow('Weird Shape', mask)

# Combine the original image with the mask
combined = cv.bitwise_and(img1, img1, mask=mask[:,:,0])

# Display the combined image
cv.imshow('Combined Result', combined)

# Wait for a key press indefinitely or for a specified amount of time (0 means indefinitely)
cv.waitKey(0)

# Destroy all the windows
cv.destroyAllWindows()
```
