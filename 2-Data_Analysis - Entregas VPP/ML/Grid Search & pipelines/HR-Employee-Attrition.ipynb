```python
import pandas as pd
```

```python
df = pd.get_dummies(df, drop_first=True)

```

```python
X = df.drop('Attrition_Yes', axis=1)
y = df['Attrition_Yes']
```

```python
y.value_counts()   # Si podemos observar están desbalanceadas, se podría pensar en un (class_weight = "balanced")
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report, roc_curve, auc
```

```python
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("classifier", LogisticRegression())  # This will be replaced in the grid search
])

# Define the parameter grid with fewer parameters
param_grid = [
    {'classifier': [LogisticRegression(class_weight="balanced")], 'classifier__C': [0.1, 1, 10]},
    {'classifier': [SVC()], 'classifier__kernel': ['linear', 'rbf'], 'classifier__C': [0.1, 1]},
    {'classifier': [XGBClassifier()], 'classifier__n_estimators': [50, 100], 'classifier__learning_rate': [0.1, 0.2]},
    {'classifier': [RandomForestClassifier()], 'classifier__n_estimators': [50, 100], 'classifier__max_depth': [None, 10]},
    {'classifier': [MLPClassifier(max_iter=1000)], 'classifier__hidden_layer_sizes': [(50,), (100,)], 'classifier__alpha': [0.0001, 0.001]}
]

# Perform grid search
grid_search = GridSearchCV(pipe, param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
print("Best Model:", best_model)
print("Best Parameters:", grid_search.best_params_)
```

```python
y_pred = best_model.predict(X_test)
# y_pred_proba = best_model.predict_proba(X_test)[:, 1]
```

```python
# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
# print("AUC-ROC:", roc_auc)
```

```python
print(classification_report(y_test, y_pred))
```
