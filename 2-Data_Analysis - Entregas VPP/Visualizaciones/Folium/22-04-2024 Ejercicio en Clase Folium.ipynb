```python
import pandas as pd
import requests

import folium # pip install folium
from folium import plugins
import ipywidgets
import geocoder # pip install geocoder
import geopy # pip install geopy
import numpy as np
import pandas as pd
from vega_datasets import data as vds # pip install vega_datasets
```

```python
maraton = {
  "type": "FeatureCollection",
  "features": [
    {
      "type": "Feature",
      "properties": {},
      "geometry": {
        "coordinates": [
          [
            -0.38292548320396236,
            39.481483902377136
          ],
          [
            -0.3752696738414443,
            39.48154351677988
          ],
          [
            -0.3664023952924538,
            39.47364686072416
          ],
          [
            -0.35957687100022895,
            39.462991160126705
          ],
          [
            -0.34834998213204926,
            39.45525989513402
          ]
        ],
        "type": "LineString"
      }
    }
  ]
}
```

```python
maraton["features"][0]["geometry"]["coordinates"][0]
```

```python
longitudes = []
latitudes = []

for coordenada in maraton["features"][0]["geometry"]["coordinates"]:
    longitud, latitud = coordenada
    longitudes.append(longitud)
    latitudes.append(latitud)


print("Longitudes:", longitudes)
print("Latitudes:", latitudes)
```

```python
tabla_maraton =  {
            "Pais":["España"]*len(latitudes),
            "Latitud" : latitudes,
            "Longitud": longitudes,
            "Punto": ["A","B","C","D","E"] 
            }

df = pd.DataFrame (tabla_maraton)
df
```

```python
mapa = folium.Map(location=[39.481457317138016, -0.3471848123771871], zoom_start= 13)

for (index,row) in df.iterrows():
    folium.Marker( location = [row.loc['Latitud'], row.loc['Longitud']],
        popup = row.loc['Punto'],
        tooltip = 'click').add_to(mapa)
mapa
```

```python
puntos_agua = ["C", "E"]


for row in df.itertuples():
    if row.Punto in puntos_agua:
        icono = folium.Icon(icon='tint', prefix='fa', color='blue') 
    else:
        icono = folium.Icon(icon='fa-person-running', prefix='fa', color='red')

    
    folium.Marker(
        location=[getattr(row, 'Latitud'), getattr(row, 'Longitud')],
        popup=getattr(row, 'Punto'),
        icon=icono
    ).add_to(mapa)

mapa
```

```python
ruta = list(zip(latitudes, longitudes))
folium.PolyLine([ruta]).add_to(mapa)
mapa
```

```python
plugins.AntPath([ruta]).add_to(mapa)
mapa
```
