```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# https://www.eldiario.es/comunitat-valenciana/produccion-residuos-urbanos-area-metropolitana-valencia-consolida-linea-bajada-pandemia_1_9884128.html
# https://emtre.es/es/entidad/noticias/valencia-y-su-area-metropolitana-generaron-650000-toneladas-de-residuos-en-2019
# https://valenciaplaza.com/valencia-y-su-area-metropolitana-reducen-los-residuos-generados-en-el-2020-en-un-4#:~:text=VAL%C3%88NCIA.,Tratamiento%20de%20Residuos%20(Emtre).
```

```python
df = pd.DataFrame({ "Año": [2019,2020,2021], 
                   "Residuos":[651401,557102, 552019]
}) 
```

```python
# plt.axhline(y=media, color='r', linestyle='-')

media = df["Residuos"].mean()
media
```

```python
df.info()
```

```python
plt.figure(figsize=(8, 6))
bars = plt.bar(df["Año"], df["Residuos"])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), va='bottom', ha='center')

# plt.axhline(y=media, color='red', linestyle='--', label='Media')

# plt.legend()
plt.xlabel('Año')
plt.ylabel('Residuos(ton)') 
plt.title('Residuos Valencia Área Metropolitana')
plt.xticks(df["Año"])
plt.show()
```

```python
colors = sns.color_palette("muted", n_colors=len(df)).as_hex()


plt.figure(figsize=(8, 4))
plt.bar(df["Año"], df["Residuos"], color=colors)
plt.title('Gestión de Residuos Área Metropolitana Valencia')
plt.xlabel('Año')
plt.ylabel('(ton)')
plt.xticks(df["Año"]) 
plt.show()
```
