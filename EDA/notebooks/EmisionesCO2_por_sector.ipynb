```python
plt.figure(figsize=(10, 6))

df_long = df.melt(id_vars=['Año'], var_name='Category', value_name='Waste Generation')
sns.lineplot(data=df_long, x='Año', y='Waste Generation', hue='Category', marker='o')

# Add title and labels
plt.title('Trends in Waste Generation Over Time')
plt.xlabel('Year')
plt.ylabel('Waste Generation')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
df_2022 = df[df['Año'] == 2022]
df_2022
```

```python
categories = df_2022.columns[1:].tolist()
sizes = df_2022.iloc[0, 1:].tolist()


plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=categories, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('muted'))


centre_circle = plt.Circle((0, 0), 0.7, color='white', fc='white', linewidth=1.25)
plt.gca().add_artist(centre_circle)
plt.title('Generación GEI por Sector 2022 (%)')


plt.axis('equal')
plt.show()
```
