```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# https://climatica.coop/10-empresas-56-emisiones-2020-repsol/


top = pd.read_csv("..\Data\Top10empreasCO2.csv",sep= ",")
top
```

```python
plt.figure(figsize=(10, 8))
bar_plot = sns.barplot(x='TONELADAS DE CO2eq.', y='EMPRESAS', data=top, palette='pastel')
plt.title('Top 10 Empresas Españolas Emisión de CO2 ')
plt.xlabel('Emisiones CO2 (millones de toneladas)')
plt.ylabel('Empresas')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_width(), '.1f'), 
                       (p.get_width(), p.get_y() + p.get_height() / 2), 
                       ha = 'left', 
                       va = 'center', 
                       xytext = (5, 0), 
                       textcoords = 'offset points') 
    
plt.show()
```
