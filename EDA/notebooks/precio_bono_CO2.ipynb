```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("..\Data\historico-precios-CO2-_2024_.csv", sep=";")
df
```

```python
df.info()
```

```python
# Cambiamos a formato fecha la columna "Fecha"

df["Fecha"] = pd.to_datetime(df['Fecha'],format='%d-%m-%Y')
```

```python
# Tuve que unir los csv, así que los hice por concat DataFrame
```

```python
df2 = pd.read_csv("..\Data\historico-precios-CO2-_2023_.csv", sep=";")
```

```python
df2["Fecha"] = pd.to_datetime(df2["Fecha"],format="%d-%m-%Y")
```

```python
df3 = pd.read_csv("..\Data\historico-precios-CO2-_2022_.csv", sep=";")
```

```python
df3["Fecha"] = pd.to_datetime(df3["Fecha"],format="%d-%m-%Y")
```

```python
df4 = pd.read_csv("..\Data\historico-precios-CO2-_2021_.csv", sep=";")
```

```python
df4["Fecha"] = pd.to_datetime(df4["Fecha"],format="%d-%m-%Y")
```

```python
df5 = pd.read_csv("..\Data\historico-precios-CO2-_2020_.csv", sep=";")
```

```python
df5["Fecha"] = pd.to_datetime(df5["Fecha"],format="%d-%m-%Y")
```

```python
union_df = pd.concat([df,df2,df3,df4,df5])
```

```python
media = df["EUA"].mean()
media
```

```python
plt.figure(figsize=(10, 6))

plt.axhline(y=media, color='red', linestyle='--', label='Media')
plt.legend()

sns.lineplot(x='Fecha', y='EUA', data=union_df, markers=True)
plt.title('Precio Bonos CO2 (2020-2024)')
plt.xlabel('Fecha')
plt.ylabel('Precios Bono CO2 (EUR)')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
