```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time


url = 'https://emtre.es/es/ecoparques/ecoparques-fijos'
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
response = requests.get(url, headers=headers)
```

```python
print(response)
```

```python
html = response.content
```

```python
print(html)
```

```python
soup = bs(html,"lxml")
```

```python
soup
```

```python
name = soup.find_all("div", class_="views-field views-field-title")
```

```python
names = []

for element in name:
    name = element.get_text(strip=True)
    names.append(name)

names
```

```python
addres = soup.find_all("div", class_="views-field views-field-field-ecoparque-direccion")
```

```python
direcciones = []

for element in addres:
    direccion = element.get_text(strip=True)
    direcciones.append(direccion)

direcciones
```

```python
locations = soup.find_all("div", class_="views-field views-field-field-ecoparque-direccion")
```

```python
geoposiciones = []

for element in addres:
    geopo = element.a.attrs["href"]
    geoposiciones.append(geopo)

geoposiciones
```

```python
data = pd.DataFrame({"Ecoparques":names,
                     "Direccion":direcciones,
                     "Geolocalizacion":geoposiciones})

df = data
df
```
