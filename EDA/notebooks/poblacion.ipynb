```python
sizes = comunidad[['P. de Alicante', 'P. de Castellón', 'P. de Valencia']].iloc[0].values
labels = ['P. de Alicante', 'P. de Castellón', 'P. de Valencia']


fig, ax = plt.subplots()
colors = ['#ff9999', '#66b3ff', '#99ff99'] 
sns.color_palette("crest", as_cmap=True)


ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, pctdistance=0.85)


centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)


ax.axis('equal')  
plt.tight_layout()
```

```python

df_2021 = pd.DataFrame(data)
sizes_2021 = df_2021.iloc[0].values
labels = df_2021.columns


palette = sns.color_palette("muted", n_colors=3)  
colors = palette.as_hex()  

# Create pie chart
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes_2021, labels=labels, colors=colors,
                                  autopct=lambda p: '{:,}'.format(int(p * sum(sizes_2021) / 100)), 
                                  startangle=90, pctdistance=0.85)


centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)


ax.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
data = {
    'P. de Alicante': [1881762],
    'P. de Castellón': [587064],
    'P. de Valencia': [2589312]
}

# Create a DataFrame for the year 2021
df_2021 = pd.DataFrame(data)

# Get population numbers for 2021
sizes_2021 = df_2021.iloc[0].values
labels = df_2021.columns
colors = ['#ff9999', '#66b3ff', '#99ff99']  # Optional: specify custom colors

# Create pie chart
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes_2021, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, pctdistance=0.85)

# Draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')  
plt.tight_layout()
plt.show()
```
