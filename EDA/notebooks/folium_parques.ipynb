```python
tabla["Latitud"] = Latitudes
tabla["Longitud"] = Longitudes
```

```python
tabla.drop(columns="Geolocalizacion", inplace=True)
```

```python
tabla
```

```python
mapa = folium.Map(location=[39.481457317138016, -0.3471848123771871], zoom_start= 11)

for (index,row) in tabla.iterrows():
    folium.Marker( location = [row.loc['Latitud'], row.loc['Longitud']],
        popup = row.loc['Ecoparques'],
        icon=folium.Icon(color='green', icon="fa fa-recycle" , prefix='fa')).add_to(mapa)
    
mapa
```

```python
mapa.save('mapa_ecoparques.html')
```
