```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
os.listdir()
```

```python
eur =  pd.read_csv("..\Data\Tabla_limpia_Residuos_Europa2", index_col=0)
eur
```

```python
# Elijo paises que estén por encima de los 50 millones de toneladas

eur = eur[eur.loc[:,"Media"] >50000000]
eur
```

```python
eur_sorted = eur.sort_values(by="Media", ascending=True)
```

```python
sns.barplot(x="Media", y="Country", data=eur_sorted, palette="muted")

plt.title("Generación Residos Europa")
plt.ylabel("")
plt.xlabel("Millones/ton")
plt.show()
```

```python
spain = eur.loc[eur["Country"] == "Spain"]
spain
```

```python
spain_long = spain.melt(id_vars=['Country'], var_name='Year', value_name='Value')
```

```python
# Creando una gráfica de línea
sns.set(style="whitegrid") 
sns.lineplot(data=spain_long, x='Year', y='Value', hue='Country', marker='o')
plt.title('Generación de Residuos en España (2010-2020)')
plt.ylabel("Miles de Millones ton Residuos ")
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor visualización
plt.show()
```
