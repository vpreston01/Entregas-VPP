```python
import pandas as pd
import numpy as np
import os
```

```python
europa = pd.read_csv("..\Data\env_wasgen_page_tabular (1).tsv", sep="\t")
```

```python
europa= europa.rename(columns={"freq,unit,hazard,nace_r2,waste,geo\TIME_PERIOD":"Prefix"})
```

```python
europa
```

```python
europa.loc[:,"Prefix"] =europa.loc[:,"Prefix"].str[-2:]
```

```python
# El DATASET vino sin el nombre de los paises, solo con prefijos. Creo una lista y creo una columna nueva estableciendo como index

nombres_pais = ["Austria", "Bosnia and Herzegovina","Belgium","Bulgaria","Cyprus","Czechia","Germany","Denmark","Estonia","Greece","Spain","Eur_27","Eu_28","Finland","France","Croatia","Hungry","Ireland","Iceland","Italy","Liechtenstein","Lithuania","Luxemburg","Latvia","Montenegro","North Macedonia","Malta","Netherlands","Norway","Poland","Portugal","Romania","Serbia","Sweden","Slovenia","Slovakia","Turkey","United Kingdom","Kosovo"]
```

```python
europa["Country"] = nombres_pais
europa.set_index("Country",inplace=True)
```

```python
# Quito el multiindex y actualizo el dataframe

europa = europa.reset_index()
```

```python
europa.head(3)
```

```python
# Antes de avanzar necesito quitar un espacio en blanco que tienen las columnas de los años

europa.columns = europa.columns.str.strip()
```

```python
# Creo una lista donde voy a iterar, no pude seleccionar las columnas con un rango.

columnas_limpiar = ["2004","2006","2008","2010","2012","2014","2016","2018","2020"]
```

```python
# convierto todos las observaciones en valores numéricos. Ojo me quedan en formato de Notación Científica, en los próximos pasos lo modifico

for columna in columnas_limpiar:
    europa[columna] =pd.to_numeric(europa[columna],errors='coerce')
```

```python
# Los NAN los convierto en cero

europa[columnas_limpiar]=europa[columnas_limpiar].fillna(0)
```

```python
# Los transformo de FLOAT (N. cientifica) a Int

for columna in columnas_limpiar:
    europa[columna] = europa[columna].astype("int")
```

```python
# Elimino la columna de los Prefijos que no me aporta nada

europa= europa.drop("Prefix",axis="columns")
```

```python
# Reviso cómo me quedaron los datos, si pasaron de Object a Int

europa.info()
```

```python
# Elimino la sumatoria de Europa y UK

filas_eliminar = europa.loc[:,"Country"].isin(["Eur_27","Eu_28","United Kingdom"])
europa =europa[-filas_eliminar]
```

```python
europa = europa.replace(0, np.nan)
```
