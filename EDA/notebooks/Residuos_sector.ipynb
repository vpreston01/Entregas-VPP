```python
import matplotlib.pyplot as plt

# Data
labels = ['Construcción', 'Gestión de residuos', 'Industria', 'Sector pecuario', 'Servicios']
sizes = [37064200, 26689800, 17255800, 6344400, 5406100]

# Create a pie chart
plt.figure(figsize=(8, 10))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3), textprops={'fontsize': 10})
plt.title('Residuos Generados por Sector')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Draw a white circle at the center to create the donut chart
centre_circle = plt.Circle((0, 0), 0.7, color='white', linewidth=0)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Construccion', 'Gestión de residuos', 'Industria', 'Sector pecuario', 'Servicios']
sizes_2019 = [35043800, 23930900, 38810600, 61840, 6496800]
sizes_2020 = [32542600, 21941500, 16450500, 6330700, 5849800]
sizes_2021 = [37064200, 26689800, 17255800, 6344400, 5406100]

# Colors
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

# Create subplots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Create pie chart for 2019
ax[0].pie(sizes_2019, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.3), textprops={'fontsize': 8})
ax[0].set_title('2019')

# Create pie chart for 2020
ax[1].pie(sizes_2020, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.3), textprops={'fontsize': 8})
ax[1].set_title('2020')

# Create pie chart for 2021
ax[2].pie(sizes_2021, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.3), textprops={'fontsize': 8})
ax[2].set_title('2021')

# Draw a white circle at the center of each pie chart to create the donut effect
for a in ax:
    centre_circle = plt.Circle((0, 0), 0.7, color='white', linewidth=0)
    a.add_artist(centre_circle)

# Display the chart
plt.show()
```

```python
labels = ['Construccion', 'Gestión de residuos', 'Industria', 'Sector pecuario', 'Servicios']
sizes_2019 = [35043800, 23930900, 38810600, 61840, 6496800]
sizes_2020 = [32542600, 21941500, 16450500, 6330700, 5849800]
sizes_2021 = [37064200, 26689800, 17255800, 6344400, 5406100]

# Generate colors from Seaborn's "muted" palette
colors = sns.color_palette("muted", n_colors=5).as_hex()

# Create subplots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Create pie chart for 2019
ax[0].pie(sizes_2019, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.3), textprops={'fontsize': 8})
ax[0].set_title('2019')

# Create pie chart for 2020
ax[1].pie(sizes_2020, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.3), textprops={'fontsize': 8})
ax[1].set_title('2020')

# Create pie chart for 2021
ax[2].pie(sizes_2021, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.3), textprops={'fontsize': 8})
ax[2].set_title('2021')

# Draw a white circle at the center of each pie chart to create the donut effect
for a in ax:
    centre_circle = plt.Circle((0, 0), 0.7, color='white', linewidth=0)
    a.add_artist(centre_circle)

# Display the chart
plt.show()
```
