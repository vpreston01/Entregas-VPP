### TEMA

Valencia es la **Capital Verde Europea 2024**, pero presenta un problema en la gestión de residuos, que junto con el encarecimiento progresivo de la energía eléctrica hace muy interesante un proyecto donde se brinde una solución para ambos problemas: una planta de **Waste to Energy (W2E)**.

La planta de Waste to Energy permite el aprovechamiento de residuos para transformarlos en energía y reciclar subproductos, de forma altamente eficiente, *solucionando la acumulación de residuos en vertederos*.

Se trata de una estructura modular donde se reúnen diferentes sistemas que permiten la separación de los residuos dependiendo del tipo y los procesa de manera diferente, obteniendo:

1. **Residuos Orgánicos**:  
   A través de un biodigestor se producen tres procesos:
   - Desde sólidos se genera un compost (fertilizante).
   - Biogas (Combustible energético).
   - Aguas residuales (lixiviados) se trata y se usan como agua no apta para consumo humano.

2. **Plásticos - Papel/Cartón - Vidrios - Escombros**  
   Separación, trituración, limpieza, compactación y venta de reciclados.

3. **Residuos Peligrosos**:
   - Material Sanitario/Industriales:
     Clasificación para incineración o encapsulamiento. 

La generación energética es posible a través de incineración (obteniendo energía eléctrica y térmica), y a través de biogás que mueve un generador.

Etcétera, ya que su estructura modular permite añadir el procesamiento de otros residuos adicionales. 

La diferencia del uso de esta tecnología respecto a las ya utilizadas habitualmente es por un lado la **termovalorización**, es decir, la obtención de energía por incineración y posterior tratamiento de los humos, y por otro la centralización de todos los procesos permitiendo la optimización de recursos evitando la acumulación en vertederos y, por último, el reciclaje. Aportando por tanto a la economía circular.
